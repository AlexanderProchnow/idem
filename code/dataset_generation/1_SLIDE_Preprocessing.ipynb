# SLIDE Preprocessing

```python
import pandas as pd
from better_profanity import profanity
```

```python
# Read from the IBM Slide dataset
df = pd.read_csv('../SLIDE_dataset/idiomLexicon.tsv',sep='\t', header=0)
df
```

```python
# Filter out the idioms that are considered ambiguous
ambiguous = df.loc[df['FilterOut(X)']== 'X'] 
```

```python
ambiguous
```

```python
df = df.drop(ambiguous.index)
df
```

```python
df.reset_index(drop=True, inplace=True)
df = df.drop(['FilterOut(X)'], axis = 1)
df
```

```python
# This is our list of idioms that we will use
idioms = df["Idiom"]
idioms
```

```python
ambiguous_idioms = ambiguous['Idiom']
ambiguous_idioms.reset_index(drop=True, inplace=True)
ambiguous_idioms
```

```python
# Some idioms in the list contain profanity, and ChatGPT-4 is refusing to accept them as input
index_list = []
for i in range(len(idioms)):              
    if(profanity.contains_profanity(idioms[i])):
        index_list.append(i)
        print(idioms[i])
```

```python
# So we filter them out of our list
idioms = idioms.drop(index_list)
```

```python
idioms.reset_index(drop=True, inplace=True)
idioms
```

```python
idioms.to_csv('idioms.csv')
```
