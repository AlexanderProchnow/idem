# 4. Evaluation

```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score
```

```python
# Load the 1000 sentences dataset from Expert F
df_eval_F = pd.read_csv('evaluations/sentence_eval_annotator1.csv.csv', index_col = 0)
df_eval_F = df_eval_F.drop(df_eval_F.columns[8], axis=1)
df_eval_F
```

```python
# Load the 100 sentences dataset from Expert 1
df_eval_F_100 = pd.read_csv("sentence_eval_100_F.csv", index_col = 0, skiprows = 4,)
df_eval_F_100
```

```python
# Combine them into a dataframe of 1100 entries, and then remove the duplicates
combined_df_F = pd.concat([df_eval_F_100, df_eval_F])
combined_df_F = combined_df_F.drop_duplicates(['index'])
combined_df_F = combined_df_F.reset_index(drop=True)
combined_df_F
```

```python
# Load the 1000 sentences dataset from expert 2
df_eval_C = pd.read_csv("evaluations/sentence_eval_annotator2", index_col = 0)
df_eval_C
```

```python
# Load the 100 sentences dataset from Expert C
df_eval_C_100 = pd.read_csv("sentence_eval_100_C.csv", index_col = 0, skiprows = 4,)
df_eval_C_100
```

```python
# Combine them into a dataframe of 1100 entries for Expert C, and then remove the duplicates
combined_df_C = pd.concat([df_eval_C_100, df_eval_C])
combined_df_C = combined_df_C.drop_duplicates(['index'])
combined_df_C = combined_df_C.reset_index(drop=True)
combined_df_C
```

```python
# Find out how many irregularities, and which ones, are contained in the dataset of Expert C
sum_inc_sent_C = 0
sum_inc_idiom_C = 0
sum_wrong_emot_C = 0
sum_inc_C = 0
sum_emot_present_C = 0

for i in range(len(df_eval_C)):
    if combined_df_C['Incorrect Sentence'].iloc[i] == 1:
        sum_inc_sent_C +=1
    if combined_df_C['Incorrect Idiom Usage'].iloc[i] == 1:
        sum_inc_idiom_C +=1
    if combined_df_C['Wrong Emotion Label'].iloc[i] == 1:
        sum_wrong_emot_C +=1
    if combined_df_C['Emotion Label Present in Sentence'].iloc[i] == 1:
        sum_emot_present_C +=1    
    if combined_df_C['Incorrect Sentence'].iloc[i] == 1 or combined_df_C['Incorrect Idiom Usage'].iloc[i] == 1 or combined_df_C['Wrong Emotion Label'].iloc[i] == 1 or combined_df_C['Emotion Label Present in Sentence'].iloc[i] == 1:
        sum_inc_C +=1
```

```python
# Get the percentages for expert C
perc_inc_sent_C = sum_inc_sent_C / len(combined_df_C)
perc_inc_idiom_C = sum_inc_idiom_C / len(combined_df_C)
perc_wrong_emot_C = sum_wrong_emot_C / len(combined_df_C)
perc_inc_C = sum_inc_C / len(combined_df_C)


print("For expert C the percentage of incorrect sentences is: ", perc_inc_sent_C*100,'%')
print("For expert C the percentage of incorrect idiom usage is: ", perc_inc_idiom_C*100,'%')
print("For expert C the percentage of wrong emotion label is: ", perc_wrong_emot_C*100,'%')
print("For expert C the percentage of sentences containing irregularities is: ", perc_inc_C*100,'%')
```

```python
# Find out how many irregularities, and which ones, are contained in the dataset of expert F
sum_inc_sent_F = 0
sum_inc_idiom_F = 0
sum_wrong_emot_F = 0
sum_inc_F = 0
sum_emot_present_F = 0

for i in range(len(combined_df_F)):
    if combined_df_F['Incorrect Sentence'].iloc[i] == 1:
        sum_inc_sent_F +=1
    if combined_df_F['Incorrect Idiom Usage'].iloc[i] == 1:
        sum_inc_idiom_F +=1
    if combined_df_F['Wrong Emotion Label'].iloc[i] == 1:
        sum_wrong_emot_F +=1
    if combined_df_F['Emotion Label Present in Sentence'].iloc[i] == 1:
        sum_emot_present_F +=1    
    if combined_df_F['Incorrect Sentence'].iloc[i] == 1 or combined_df_F['Incorrect Idiom Usage'].iloc[i] == 1 or combined_df_F['Wrong Emotion Label'].iloc[i] == 1 or combined_df_F['Emotion Label Present in Sentence'].iloc[i] == 1:
        sum_inc_F +=1
```

```python
# Get the percentages for expert F
perc_inc_sent_F = sum_inc_sent_F / len(combined_df_F)
perc_inc_idiom_F = sum_inc_idiom_F / len(combined_df_F)
perc_wrong_emot_F = sum_wrong_emot_F / len(combined_df_F)
perc_inc_F = sum_inc_F / len(combined_df_F)


print("For expert F the percentage of incorrect sentences is: ", perc_inc_sent_F*100,'%')
print("For expert F the percentage of incorrect idiom usage is: ", perc_inc_idiom_F*100,'%')
print("For expert F the percentage of wrong emotion label is: ", perc_wrong_emot_F*100,'%')
print("For expert F the percentage of sentences containing irregularities is: ", perc_inc_F*100,'%')
```

```python
# Investigate the level of agreement between the two experts
agr_inc_sent = 0
agr_inc_idiom = 0
agr_wrong_emot = 0
agr_inc = 0
agr_emot_pres = 0

for i in range(len(df_eval_F)):
    if combined_df_F['Incorrect Sentence'].iloc[i] == combined_df_C['Incorrect Sentence'].iloc[i]:
        agr_inc_sent +=1
    if combined_df_F['Incorrect Idiom Usage'].iloc[i] == combined_df_C['Incorrect Idiom Usage'].iloc[i]:
        agr_inc_idiom +=1
    if combined_df_F['Wrong Emotion Label'].iloc[i] == combined_df_C['Wrong Emotion Label'].iloc[i]:
        agr_wrong_emot +=1
    if combined_df_F['Emotion Label Present in Sentence'].iloc[i] == combined_df_C['Emotion Label Present in Sentence'].iloc[i]:
        agr_emot_pres +=1
        
agr_inc = agr_inc_sent + agr_inc_idiom + agr_wrong_emot + agr_emot_pres
```

```python
# Get the percentages of agreement
perc_agr_inc_sent= agr_inc_sent / len(combined_df_F)
perc_agr_inc_idiom = agr_inc_idiom / len(combined_df_F)
perc_agr_wrong_emot = agr_wrong_emot  / len(combined_df_F)
perc_agr_emot_pres = agr_emot_pres  / len(combined_df_F)

perc_agr_inc = agr_inc / (4*len(combined_df_F))

print("The percentage of agreement of experts for incorrect sentences is: ", perc_agr_inc_sent*100,'%')
print("The percentage of agreement of experts for incorrect idiom usage is: ", perc_agr_inc_idiom*100,'%')
print("The percentage of agreement of experts for wrong emotion label is: ", perc_agr_wrong_emot*100,'%')
print("The percentage of agreement of experts for emotion label present in the sentence is: ", perc_agr_wrong_emot*100,'%')
print("The overall percentage of agreement is: ", perc_agr_inc*100,'%')
```

```python
# Measure the Kappa-Cohen Agreement Score
column1_C = combined_df_C['Incorrect Sentence'].tolist()
column2_C = combined_df_C['Incorrect Idiom Usage'].tolist()
column3_C = combined_df_C['Wrong Emotion Label'].tolist()
column4_C = combined_df_C['Emotion Label Present in Sentence'].tolist()

# Combine the four lists into one list
combined_list_C = column1_C + column2_C + column3_C + column4_C

column1_F = combined_df_F['Incorrect Sentence'].tolist()
column2_F = combined_df_F['Incorrect Idiom Usage'].tolist()
column3_F = combined_df_F['Wrong Emotion Label'].tolist()
column4_F = combined_df_F['Emotion Label Present in Sentence'].tolist()

# Combine the four lists into one list
combined_list_F = column1_F + column2_F + column3_F + column4_F
```

```python
kappa_cohen_score = cohen_kappa_score(combined_list_C, combined_list_F)
print('Kappa-Cohen Score of agreement between judges is', kappa_cohen_score)
```
