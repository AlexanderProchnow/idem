# GPT-4 Experiments

Experiment 1: no idiom info  
Experiment 2: idiom info  
Experiment 3: idiom definition  
Experiment 4: idiom sentiment  
Experiment 5: both infos  
Experiment 6: demonstrations (prompt-based learning) using 5 examples. Shown in ["Prompt Programming for Large Language Models: Beyond the Few-Shot Paradigm"](https://arxiv.org/abs/2102.07350) to improve the model's understanding of the task in some cases.


## Results


|                     |Accuracy|  F1
|---------------------|--------|------ 
|No idiom information | 0.610  | 0.611 
|Idiom information    | 0.585  | 0.584
|Definition           | 0.556  | 0.557 
|Sentiment            | 0.555  | 0.559
|Both Infos           | 0.563  | 0.566
|Demonstrations (n=5) | 0.591  | 0.585

```python
import pandas as pd
from gpt4_lib import GPT4Experiment
```

```python
experiments = ['no_info', 'idiom_info', 'idiom_def', 'idiom_sent', 'idiom_def_sent', 'demonstrations']
```

```python
exp = GPT4Experiment('no_info')
```

```python
exp.run()#print_prompt=True)
```

```python
exp.calculate_metrics()
```
